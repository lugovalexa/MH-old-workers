```python
# Number of children to categoric
df["nb_children"] = pd.to_numeric(df["nb_children"], errors="coerce")
bin_edges = [-1, 0, 1, 2, 4, np.inf]
bin_labels = [
    "No children",
    "1 child",
    "2 children",
    "3 or 4 children",
    "5 or more children",
]
df["nb_children_cat"] = pd.cut(
    df["nb_children"], bins=bin_edges, labels=bin_labels, include_lowest=True
)
```

```python
# Categoric variables for df
for var in ["gender", "nb_children_cat", "job_status"]:
    value_counts = df[var].value_counts(dropna=False)
    percentages = round((value_counts / len(df)) * 100, 2)
    print(pd.DataFrame({"Count": value_counts, "Percentage": percentages}))
```

```python
# Categoric variables by wave
for wave, group in df.groupby("wave"):
    print(f"Wave {wave} Statistics:")
    print(f"--------------------------------------")
    for var in ["gender", "nb_children_cat", "job_status"]:
        value_counts = group[var].value_counts(dropna=False)
        percentages = round((value_counts / len(group)) * 100, 2)
        print(pd.DataFrame({"Count": value_counts, "Percentage": percentages}))
    print(f"--------------------------------------")
```
